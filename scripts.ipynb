```python
# For example, given 5 epochs
history = {0: [np.random.uniform(low=0, high=1) * 100, np.random.uniform(low=0, high=1) * 100, np.random.uniform(low=0, high=1), np.random.uniform(low=0, high=1)],
            1: [np.random.uniform(low=0, high=1) * 100, np.random.uniform(low=0, high=1) * 100, np.random.uniform(low=0, high=1), np.random.uniform(low=0, high=1)],
            2: [np.random.uniform(low=0, high=1) * 100, np.random.uniform(low=0, high=1) * 100, np.random.uniform(low=0, high=1), np.random.uniform(low=0, high=1)],
            3: [np.random.uniform(low=0, high=1) * 100, np.random.uniform(low=0, high=1) * 100, np.random.uniform(low=0, high=1), np.random.uniform(low=0, high=1)],
            4: [np.random.uniform(low=0, high=1) * 100, np.random.uniform(low=0, high=1) * 100, np.random.uniform(low=0, high=1), np.random.uniform(low=0, high=1)]}
history
```

```python
def plot_figures(history, info='accuracy'):
    x_positions = history.keys()

    # Grab the train_acc_history
    train_acc_history = [v[0] for k, v in history.items()]

    # Grab the val_acc_history
    val_acc_history = [v[1] for k, v in history.items()]

    # Grab the train_loss_history
    train_loss_history = [v[2] for k, v in history.items()]

    # Grab the val_loss_history
    val_loss_history = [v[3] for k, v in history.items()]

    if info == 'loss':
        plt.plot(x_positions, train_loss_history, label='Train')
        plt.plot(x_positions, val_loss_history, label='Validation')
        plt.xlabel('Epoch')
        plt.ylabel('%')
        plt.legend(loc='upper right')
        plt.title('Loss')
        plt.savefig('loss.png', bbox_inches='tight')
    else:
        plt.plot(x_positions, train_acc_history, label='Train')
        plt.plot(x_positions, val_acc_history, label='Validation')
        plt.xlabel('Epoch')
        plt.ylabel('%')
        plt.legend(loc='upper right')
        plt.title('Accuracy')
        plt.savefig('accuracy.png', bbox_inches='tight')
```

```python
plot_figures(history, info='accuracy')
```

```python
plot_figures(history, info='loss')
```
